```python
import numpy as np
import collections
import scipy.special
```

```python
def compute_count_expected(n_r, n_g, n_b, n_1r, n_1g, n_1b, n_2):
    comb = scipy.special.comb
    n = n_r + n_g + n_b
    n_1 = n_1r + n_1g + n_1b
    return int(comb(n_r, n_1r) * comb(n_g, n_1g) * comb(n_b, n_1b) * comb(n - n_1, n_2))
```

```python
n_r = 3; n_g = 5; n_b = 4
n = n_r + n_g + n_b

arange = np.arange(3**n, dtype=np.int64)
index = []
for i in range(n):
    index.append(arange % 3)
    arange = arange // 3
index = np.stack(index)

n_1r, n_2r, n_3r = (np.sum(index[:n_r] == a, axis=0) for a in range(3))
n_1g, n_2g, n_3g = (np.sum(index[n_r:n_r + n_g] == a, axis=0) for a in range(3))
n_1b, n_2b, n_3b = (np.sum(index[n_r + n_g:] == a, axis=0) for a in range(3))
n_2 = n_2r + n_2g + n_2b

key = np.stack([n_1r, n_1g, n_1b, n_2], axis=1)
key = [tuple(item.tolist()) for item in key]
key_count = collections.Counter(key)

for (n_1r, n_1g, n_1b, n_2), count in key_count.items():
    count_expected = compute_count_expected(n_r, n_g, n_b, n_1r, n_1g, n_1b, n_2)
    if count != count_expected:
        print(f"({n_1r}, {n_1g}, {n_1b}, {n_2}): {count} vs {count_expected}")
```
